## Points to note - 1
List some of the features that would be inefficient in predicting tomorrow's rainfall. There will be a question in the quiz that follows based on this observation.


When predicting "RainTomorrow" (whether there will be at least 1mm of rain tomorrow) using the dataset you provided, some features might be less efficient—meaning they contribute minimal predictive power, are redundant, or have weak direct influence on the outcome. Efficiency here refers to how much a feature improves accuracy relative to the complexity it adds to a model. Based on the dataset’s columns and their descriptions, here are some features that could be inefficient for predicting tomorrow’s rainfall, along with reasoning:
Potentially Inefficient Features

    Sunshine (Amount of bright sunshine, hours):
        Why Inefficient: Sunshine reflects current-day conditions and indirectly affects moisture via evaporation, but its impact on tomorrow’s rain is less direct. Rainfall is more strongly tied to humidity, pressure, and wind patterns. Sunshine might correlate inversely with cloud cover, but it’s likely redundant when Cloud9am and Cloud3pm are available.
        Caveat: It could matter in specific climates (e.g., where strong diurnal heating triggers convection), but this is niche.
    Evaporation (Amount of evaporation, Millimeters):
        Why Inefficient: Evaporation measures moisture leaving the surface today, which might influence local humidity over time. However, tomorrow’s rain depends more on moisture already in the atmosphere (captured by Humidity9am/3pm) or advected by wind. Its effect is too lagged and indirect for a one-day-ahead prediction.
        Caveat: It might help in arid regions where surface moisture is a limiting factor, but this is context-specific.
    WindGustDir (Direction of the strongest gust, Compass Points):
        Why Inefficient: The direction of the strongest gust is a single, momentary event and less representative of broader wind patterns than WindDir9am or WindDir3pm (averaged over 10 minutes). It’s noisier and less stable as a predictor, especially since gusts might not align with moisture-transporting winds.
        Caveat: It could matter if gusts signal an abrupt front, but Pressure9am/3pm likely capture that better.
    WindGustSpeed (Speed of the strongest gust, Kilometers/Hour):
        Why Inefficient: Similar to WindGustDir, the speed of a single gust is a high-variance snapshot. Average wind speeds (WindSpeed9am/3pm) better reflect sustained conditions that move weather systems. Gust speed might correlate with storminess, but it’s less consistent for rain prediction.
        Caveat: High gusts might hint at instability, but this is secondary to humidity and pressure.
    Cloud9am (Fraction of sky obscured by cloud at 9am, Eights):
        Why Inefficient: While cloud cover relates to current weather, its predictive power for tomorrow is weaker than Cloud3pm (closer to the forecast period). Morning clouds might dissipate or form unpredictably by the next day, making this less reliable on its own.
        Caveat: If paired with trends (e.g., Cloud9am to Cloud3pm), it could add value, but alone it’s less efficient.
    MinTemp (Minimum temperature, Celsius):
        Why Inefficient: Minimum temperature (likely overnight) has a weaker link to tomorrow’s rain than Temp9am or Temp3pm, which reflect daytime heating and atmospheric state closer to the prediction window. It’s more a byproduct of current weather than a driver of future rain.
        Caveat: Extreme lows might signal clear skies (less rain), but this is indirect.
    MaxTemp (Maximum temperature, Celsius):
        Why Inefficient: Maximum temperature captures today’s peak heat, which might influence convection, but its effect on tomorrow’s rain is diluted compared to Temp3pm (a snapshot closer to evening). It’s also redundant with other temperature measures.
        Caveat: In convective rainfall climates (e.g., tropics), it might play a role, but it’s still secondary.

Why These Are Inefficient

    Redundancy: Features like Sunshine, MinTemp, and MaxTemp overlap with stronger predictors (e.g., Cloud3pm, Temp3pm), adding little unique information.
    Temporal Distance: Variables like Cloud9am or MinTemp are further from the "tomorrow" timeframe, reducing their relevance compared to 3pm measurements.
    Indirect Influence: Evaporation and Sunshine affect rain through slow or multi-step processes, unlike Humidity9am/3pm or Pressure9am/3pm, which directly signal rain potential.
    Noise: WindGustDir and WindGustSpeed are sporadic, making them less reliable than averaged wind metrics.

Features That Are Likely More Efficient (For Context)
To contrast, features like Humidity3pm, Pressure3pm, WindDir3pm, WindSpeed3pm, Cloud3pm, and RainToday are typically more efficient because they:

    Are closer in time to the prediction target (tomorrow).
    Directly measure moisture, pressure systems, or weather persistence, which are core drivers of rainfall.

Final Notes
Inefficiency doesn’t mean these features are useless—some might still contribute in a model (e.g., via feature interactions). However, in a practical sense, they’re less likely to be the heavy hitters. A feature selection process (e.g., correlation analysis, mutual information, or model-based importance scores) could confirm this, but based on meteorological intuition and the dataset’s structure, these stand out as less impactful for predicting "RainTomorrow."

## Extracting a seasonality feature
Now consider the `Date` column. We expect the weather patterns to be seasonal, having different predictablitiy levels in winter and summer for example.  
There may be some variation with `Year` as well, but we'll leave that out for now.
Let's engineer a `Season` feature from `Date` and drop `Date` afterward, since it is most likely less informative than season. 
An easy way to do this is to define a function that assigns seasons to given months, then use that function to transform the `Date` column.


### Create a function to map dates to seasons


````python
def date_to_season(date):
    month = date.month
    if (month == 12) or (month == 1) or (month == 2):
        return 'Summer'
    elif (month == 3) or (month == 4) or (month == 5):
        return 'Autumn'
    elif (month == 6) or (month == 7) or (month == 8):
        return 'Winter'
    elif (month == 9) or (month == 10) or (month == 11):
        return 'Spring'
````

### Define separate transformers for both feature types and combine them into a single preprocessing transformer


````python
# Scale the numeric features
numeric_transformer = Pipeline(steps=[('scaler', StandardScaler())])

# One-hot encode the categoricals 
categorical_transformer = Pipeline(steps=[('onehot', OneHotEncoder(handle_unknown='ignore'))])
````

## Exercise 7. Combine the transformers into a single preprocessing column transformer
Complete the followng code:  
```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, ...),
        ('cat', categorical_transformer, ...)
    ]
)
```


````python
# Write your response.
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ]
    
)
````

## Exercise 8. Create a pipeline by combining the preprocessing with a Random Forest classifier
Complete the following code:
```python

    ('preprocessor', ...),
    ('...', RandomForestClassifier(random_state=42))
])
```


````python
# Write your response.
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42))
])
````

Recall that to obtain the categorical feature importances, we have to work our way backward through the modelling pipeline to associate the feature importances with their original input variables, not the one-hot encoded ones.
````python
## Write your response.
feature_importances = grid_search.best_estimator_['classifier'].feature_importances_
display(feature_importances)
````

Now let's extract the feature importances and plot them as a bar graph.


````python
# Combine numeric and categorical feature names
feature_names = numerical_features + list(grid_search.best_estimator_['preprocessor']
                                        .named_transformers_['cat']
                                        .named_steps['onehot']
                                        .get_feature_names_out(categorical_features))

feature_importances = grid_search.best_estimator_['classifier'].feature_importances_

N = 20  # Change this number to display more or fewer features

importance_df = pd.DataFrame({'Feature': feature_names,
                              'Importance': feature_importances
                             }).sort_values(by='Importance', ascending=False)
importance_df.head(N)

# Plotting
plt.figure(figsize=(16, 16))
plt.barh(importance_df['Feature'], importance_df['Importance'], color='skyblue')
plt.gca().invert_yaxis()  # Invert y-axis to show the most important feature on top
plt.title(f'Top {N} Most Important Features in predicting whether it will rain today')
plt.xlabel('Importance Score')
plt.show()
````
